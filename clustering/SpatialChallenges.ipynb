```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
n = 300

position = np.linspace(0, n+1, n)


plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.facecolor'] = "w"
plt.rcParams['figure.figsize'] = 9.4, 5.5

fig, axs = plt.subplots(1, 5, sharey=True)
axs = axs.ravel()


cell1 = np.sin(position/30) + np.cos(position/51) + np.sin(position/13) + np.random.normal(0, 0.6, n)
axs[0].barh(position, 
            cell1, 
            color='#C5A866')

axs[0].set_xlabel("Cell 1", fontsize=16)
axs[0].set_yticks([])
axs[0].set_xticks([])

cell2 = np.sin(position/30) + np.cos(position/48) + np.sin(position/15) + np.random.normal(0, 0.6, n)
axs[2].barh(position, 
            cell2, 
            color='#C5A866')

axs[2].set_xlabel("Cell 3", fontsize=16)
axs[2].set_yticks([])
axs[2].set_xticks([])


cell3 = np.sin(position/75) + np.sin(position/300) + np.random.normal(0, 0.6, n)
axs[1].barh(position, 
            cell3, 
            color='#9593C7',)

axs[1].set_xlabel("Cell 2", fontsize=16)
axs[1].set_yticks([])
axs[1].set_xticks([])


spot = cell1+cell2+cell3
axs[3].barh(position, 
            spot, 
            color='k', )

axs[3].set_xlabel("Spot Signature", fontsize=16)
axs[3].set_yticks([])
axs[3].set_xticks([])

p = 0.25 # percent to retain
pDrop = int((1 - p) * n)

ind = np.random.choice(list(range(n)), pDrop, replace=False)
spot[ind] = 0

axs[4].barh(position, 
            spot, 
            color='#C03830',)

axs[4].set_xlabel("Observed Spot", fontsize=16)
axs[4].set_yticks([])
axs[4].set_xticks([])

plt.tight_layout()
```
