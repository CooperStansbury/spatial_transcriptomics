```python
cardOutDir = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDOuputs/"

dfList = []

for f in os.listdir(cardOutDir):
    if not 'macrophage' in f:
        fullPath = f"{cardOutDir}{f}"
        key = f.split("_")[0]
        df = pd.read_csv(fullPath)
        df = df.rename(columns={'Unnamed: 0' : 'spotId'})
        df['key'] = key
        dfList.append(df)

df = pd.concat(dfList, ignore_index=True)
print(f"{df.shape=}")
print(df['key'].value_counts())
print()
df.head()
```

```python
coordDir = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/"

dfList = []

for f in os.listdir(coordDir):
    if "coord" in f:
        fullPath = f"{coordDir}{f}"
        key = f.split("_")[0]
        cdf = pd.read_csv(fullPath)
        cdf = cdf.rename(columns={'Unnamed: 0' : 'spotId'})
        cdf['key'] = key
        dfList.append(cdf)

cdf = pd.concat(dfList, ignore_index=True)
print(f"{cdf.shape=}")
print(cdf['key'].value_counts())
print()
cdf.head()
```

```python
"""Merge CARD predictions with spatial coordinates"""
df = pd.merge(cdf, df,
              how='left', 
              left_on=['spotId', 'key'],
              right_on=['spotId', 'key'])

df = df.fillna(0)
df = df.drop_duplicates() # very important!
print(df['key'].value_counts())
df.head()
```

```python
def drawPieMarker(xs, ys, ratios, sizes, colors):
    assert sum(ratios) <= 1, 'sum of ratios needs to be < 1'

    markers = []
    previous = 0
    # calculate the points of the pie pieces
    for color, ratio in zip(colors, ratios):
        this = 2 * np.pi * ratio + previous
        x  = [0] + np.cos(np.linspace(previous, this, 10)).tolist() + [0]
        y  = [0] + np.sin(np.linspace(previous, this, 10)).tolist() + [0]
        xy = np.column_stack([x, y])
        previous = this
        markers.append({'marker' : xy, 
                        's' : np.abs(xy).max()**2*np.array(sizes), 
                        'facecolor' : color,
                        'linewidth' : 0.2,
                        'edgecolor' : 'k'})

    # scatter each of the pie pieces to create pies
    for marker in markers:
        ax.scatter(xs, ys, **marker)

key = 'ND'

kdf = df[df['key'] == key].reset_index(drop=True)
cols = [
    'x', 'y', 
    'B cells', 
    'Macrophages',
    'Dendritic cells',
]
kdf = kdf[cols]
kdf['other'] = 1 - kdf[['B cells', 'Macrophages','Dendritic cells']].sum(axis=1)
kdf.head()
colors = ['green', 'orange', 'purple', 'lightgray']

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.facecolor'] = "w"
plt.rcParams['figure.figsize'] = 8, 8
fig, ax = plt.subplots()


kdf = kdf[(kdf['x'] > 400) & (kdf['x'] < 900)]
kdf = kdf[(kdf['y'] > 400) & (kdf['y'] < 900)]
print(f"{kdf.shape=}")

for idx, row in kdf.iterrows():    
    x = row['x']
    y = row['y']
    ratios = row[['B cells', 'Macrophages','Dendritic cells', 'other']].values
    
    drawPieMarker(xs=x,
                  ys=y,
                  ratios=ratios,
                  sizes=250,
                  colors=colors)
    
    
ax = plt.gca()
ax.invert_yaxis()
ax.set_xticks([])
ax.set_yticks([])

# ax.scatter(kdf['x'], kdf['y'], c='k', marker=".", zorder=1)

sns.despine(left=True, bottom=True)
```
