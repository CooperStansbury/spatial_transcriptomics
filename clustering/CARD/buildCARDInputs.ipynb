```python
""" gloabl inputs """

sfPath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/scanpy/macrophageSubtypes.csv"
dirPath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/cleanData/"
outdir = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/"

sf = pd.read_csv(sfPath)
sf['cellId'] = sf['cellId'].str.replace("-", ".")


keys = ['ND', 'HFD8', 'HFD14']

annotDf = []
rnaDf = None

for key in keys:
    clusterPath = f"{dirPath}{key}_clusters.pq"
    cdf = pd.read_parquet(clusterPath, engine='fastparquet')
    cdf['sampleInfo'] = key
    cdf['cellId'] = cdf['cellId'].str.replace("-", ".")
    cdf = cdf.drop(['clusterId'], axis=1)
    
    subtypes = sf[sf['key'] == key]
    
    cdf = pd.merge(cdf, subtypes,
                   how='left',
                   left_on='cellId',
                   right_on='cellId')
    
    # cdf = cdf[cdf['cellType_x'].isin(keep)]
    cdf['cellType'] = np.where(cdf['subtypeId'].isna(), cdf['cellType_x'], cdf['subtypeId'])
    cdf = cdf[['cellId', 'cellType', 'sampleInfo']]
    # add key to cellId
    cdf['cellId'] = cdf['cellId'] + "_" + cdf['sampleInfo']
    
    
    rnaPath = f"{dirPath}{key}_rna.pq"
    rdf = pd.read_parquet(rnaPath)
    rdf.index = rdf.index.str.replace("-", ".")
    rdf.index = rdf.index.astype(str) + f"_{key}" 

    
    sptPath = f"{dirPath}{key}_spt.pq"
    sdf = pd.read_parquet(sptPath)
    sdf.index = sdf.index.str.replace("-", ".")
    sdf.index = sdf.index.astype(str) + f"_{key}" 
    
    xyPath = f"{dirPath}{key}_coords.pq"
    xydf = pd.read_parquet(xyPath)
    xydf.index = xydf.index.str.replace("-", ".")
    xydf.index = xydf.index.astype(str) + f"_{key}" 

    # remove filtered cells
    rdf = rdf[rdf.index.isin(cdf['cellId'])]
    
    # transpose the rna and spt datasets
    rdf = rdf.T
    sdf = sdf.T
    
    # filter out genes not expressed in both
    rnaGenes = list(rdf.index)
    sptGenes = list(sdf.index)
    
    rdf = rdf[rdf.index.isin(sptGenes)]
    sdf = sdf[sdf.index.isin(rnaGenes)]
    
    # structure the combined RNA dataframes
    if rnaDf is None:
        rnaDf = rdf
    else:
        rnaDf = pd.merge(rnaDf, rdf, 
                         how='left', 
                         right_index=True,
                         left_index=True)

    # structure the annotations
    annotDf.append(cdf)
    
    # save the spatial data
    sptPath = f"{outdir}{key}_global_spt.csv"
    xyPath = f"{outdir}{key}_global_coords.csv"
    
    sdf.to_csv(sptPath, index=True)
    xydf.to_csv(xyPath, index=True)
        
    print(key, cdf.shape, rdf.shape, sdf.shape, xydf.shape)
    
# combine the RNA signatures
annotDf = pd.concat(annotDf)
print(f"{rnaDf.shape=}")
print(f"{annotDf.shape=}")
    
# RNA outpath
rnaPath = f"{outdir}_global_rna.csv"
annotationPath = f"{outdir}_global_clusters.csv"

rnaDf.to_csv(rnaPath, index=True)
annotDf.to_csv(annotationPath, index=True)
print('done')
```
