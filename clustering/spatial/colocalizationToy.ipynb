```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import itertools
```

```python
# create spot grid
geneSize = 200
nbox = 2
boxsize = 4000
xbox = np.linspace(0.25, 0.75, nbox)
ybox = np.linspace(0.25, 0.75, nbox)
boxes = np.array(list(itertools.product(*[xbox, ybox])))

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 7,3.5
plt.rcParams['figure.facecolor'] = "w"

fig, axs = plt.subplots(1, 2)
axs = axs.ravel()

axs[0].scatter(boxes[:, 0], 
               boxes[:, 1], 
               alpha=0.8, 
               s=boxsize, 
               linewidth=2,
               ec='k', 
               c='lightgrey')

g1 = np.random.normal(0, 0.05, boxes.shape) + boxes
g2 = np.random.normal(0, 0.06, boxes.shape) + boxes


axs[0].scatter(g1[0:1, 0], 
               g1[0:1, 1], 
               alpha=0.8, 
               s=geneSize,
               c='r')

axs[0].scatter(g2[1:3:, 0], 
               g2[1:3:, 1], 
               alpha=0.8, 
               s=geneSize,
               c='b')

axs[0].set_xlim([0, 1])
axs[0].set_ylim([0, 1])


axs[0].set_xticks([])
axs[0].set_yticks([])


""" HIGHLY colocatlized """
axs[1].scatter(boxes[:, 0], 
               boxes[:, 1], 
               alpha=0.8, 
               s=boxsize, 
               linewidth=2,
               ec='k', 
               c='lightgrey')

g1 = np.random.normal(0, 0.05, boxes.shape) + boxes
g2 = np.random.normal(0, 0.06, boxes.shape) + boxes

axs[1].scatter(g1[:, 0], 
               g1[:, 1], 
               alpha=0.8, 
               s=geneSize,
               c='r')

axs[1].scatter(g2[:, 0], 
               g2[:, 1], 
               alpha=0.8, 
               s=geneSize,
               c='b')

axs[1].set_xlim([0, 1])
axs[1].set_ylim([0, 1])


axs[1].set_xticks([])
axs[1].set_yticks([])

# sns.despine(left=True, bottom=True)

plt.tight_layout()
```
