```python
cardOutDir = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDOuputs/"

dfList = []

for f in os.listdir(cardOutDir):
    if 'macrophage' in f:
        fullPath = f"{cardOutDir}{f}"
        key = f.split("_")[0]
        df = pd.read_csv(fullPath)
        df = df.rename(columns={'Unnamed: 0' : 'spotId'})
        df['key'] = key
        dfList.append(df)

df = pd.concat(dfList, ignore_index=True)
print(f"{df.shape=}")
print(df['key'].value_counts())
print()
df.head()
```

```python
sptDir =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/"

spt = {}

for f in os.listdir(sptDir):
    if "spt" in f and "macrophage" in f:
        print(f)
        fullPath = f"{sptDir}{f}"
        key = f.split("_")[0]
        sdf = pd.read_csv(fullPath)
        sdf = sdf.rename(columns={'Unnamed: 0' : 'gene'})
        sdf = sdf.set_index('gene')
        sdf = sdf.T
        sdf.index = sdf.index.str.replace("-", ".")
        sdf = ut.normalize(sdf, 1e6)
        spt[key] = sdf
    
print('done')
```

```python
coordDir = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/"

dfList = []

for f in os.listdir(coordDir):
    if "coord" in f:
        fullPath = f"{coordDir}{f}"
        key = f.split("_")[0]
        cdf = pd.read_csv(fullPath)
        cdf = cdf.rename(columns={'Unnamed: 0' : 'spotId'})
        cdf['key'] = key
        dfList.append(cdf)

cdf = pd.concat(dfList, ignore_index=True)
print(f"{cdf.shape=}")
print(cdf['key'].value_counts())
print()
cdf.head()
```

```python
"""Merge CARD predictions with spatial coordinates"""
df = pd.merge(cdf, df,
              how='left', 
              left_on=['spotId', 'key'],
              right_on=['spotId', 'key'])

df = df.fillna(0)
df = df.drop_duplicates() # very important!
print(df['key'].value_counts())
df.head()
```

```python
# build distance matrices
knn = 21
r = 0.95 # r % of the variance
metric = 'euclidean'

keys = ['ND', 'HFD8', 'HFD14']

simMat = {}

for key in keys:
    X = spt[key]
    geneNames = X.columns.to_list()
    pca = PCA(n_components=r)
    
    # transpose!
    pcaModel = pca.fit(X.T)
    Xpca = pca.fit_transform(X.T)
    
    # if building distance matrix, non-sparse
    A = metrics.pairwise_distances(Xpca, metric=metric, 
                                   force_all_finite=True)
    
    A = pd.DataFrame(A, columns=geneNames, index=geneNames)
    
    simMat[key] = A
    
    print(f"{X.shape=} {Xpca.shape=} {A.shape=}")
    
print('done')
```

```python
def getSimGenes(gene, simat):
    res = simat[gene]
    res = res.sort_values()
    return res
    
gene = 'TREM2'
key = 'HFD14'

getSimGenes(gene, simMat[key]).head()
```
