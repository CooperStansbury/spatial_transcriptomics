```python
def segement_image(image, unshrp_rad=100, unshrp_amnt=100, min_peak_dist=60, 
                   line=False, perc=25, blur=3, erode=5, simple=True):
    """A function to segement the cells in an image 
    
    Params:
        image (np.array): image
        unshrp_rad (int): pixel raduis for unsharp masking 
        unshrp_amnt (int): scaling for unsharp masking
        min_peak_dist (int): minimum peak distance for overlapping objects
        line (bool): if true, draw 1 pixel line separating segemented objects 
        
    Returns:
        res (np.array): the preprocessed image
        cells (np.array): the segemented cells from the processed image
    """
    res = rgb2gray(image)

    # unsharp masking for illumination correction
    res = unsharp_mask(res, radius=unshrp_rad, amount=unshrp_amnt)
    res = filters.median(res)

    # fill holes
    seed = np.copy(res)
    seed[1:-1, 1:-1] = res.max()
    mask = res
    res = reconstruction(seed, mask, method='erosion')
    
    res = filters.gaussian(res, sigma=blur)
    # hard threshold
    t = np.percentile(res, perc)
    res = np.where(res < t, 0, 1)
    
    res = morphology.erosion(res, footprint=np.ones((erode, erode)))
    res = filters.median(res)

    """method, simple or complex """
    if simple:
        thresh = res > filters.threshold_otsu(res)
        cells = measure.label(thresh)
        
    else:
        # thresholding/segmenting
        distance = ndi.distance_transform_edt(res)
        local_max_coords = feature.peak_local_max(distance, min_distance=min_peak_dist)
        local_max_mask = np.zeros(distance.shape, dtype=bool)
        local_max_mask[tuple(local_max_coords.T)] = True
        labels = measure.label(local_max_mask)

        cells = segmentation.watershed(-distance, 
                                       labels, 
                                       mask=res, 
                                       watershed_line=line)
    
    
    
    return res, cells
    
    
    
key = "B005_11_2_HFD8"
image = imgs[key]
res, cells = segement_image(image, 
                            unshrp_amnt=200,
                            simple=False)

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 12, 12    

fig, axs = plt.subplots(1, 2)
axs = axs.ravel()
    
axs[0].imshow(image, cmap='gray')
axs[0].axis(False)
axs[0].set_title(f"{key.split('_')[-1]}")

axs[1].imshow(color.label2rgb(cells))
axs[1].axis(False)
axs[1].set_title("Segmented Image")

plt.tight_layout()
```

```python
def getFeatures(cells):
    """A function to return a dataframe of features from the cells"""
    
    new_rows = []
    props = regionprops(cells)

    for region in regionprops(cells):
        
        if region.perimeter > 0:
            circularity = 4*math.pi * (region.area / region.perimeter**2)
        else: 
            circularity = 0
        
        row = {
            'area' : region.area,
            'area_convex' : region.area_convex,
            'area_filled' : region.area_filled,
            'axis_major_length' : region.axis_major_length,
            'axis_minor_length' : region.axis_minor_length, 
            'centroid' : region.centroid,
            'eccentricity' : region.eccentricity,
            'equivalent_diameter_area' : region.equivalent_diameter_area,
            'extent' : region.extent,
            'feret_diameter_max' : region.feret_diameter_max,
            'label' : region.label,
            'orientation' : region.orientation,
            'perimeter': region.perimeter,
            'perimeter_crofton': region.perimeter_crofton,
            'solidity': region.solidity,
            'circularity' : circularity,
        }
        
        new_rows.append(row)


    return pd.DataFrame(new_rows), props

df, props = getFeatures(cells)
print(df.shape)
df.head()
```

```python
plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 8, 8

def plot_cell(ax, points, lc='green', fc='green', 
              lw=0.5, area_threshold=10000):
    if len(points) > 3:
        try:
            hull = ConvexHull(points)
            
            if hull.area < area_threshold:
                ax.plot(points[hull.vertices, 1], 
                        points[hull.vertices, 0], 
                        color=lc, 
                        lw=lw, 
                        zorder=1)
                
                ax.fill(points[hull.vertices, 1], 
                        points[hull.vertices, 0], 
                        color=fc, 
                        alpha=0.4, 
                        zorder=1)
        except:
            pass
        
        
"""
TODO: remove edges
"""

A = np.ones((res.shape))

fill_pass = [(df['circularity'] > 0.1) & (df['circularity'] < 1) & (df['area'] > 200)][0]
df['INCLUDE'] = np.where(fill_pass, 1, 0)

plt.imshow(A, cmap='binary_r', zorder=0)    
ax = plt.gca()

mask = df['area'] < 50000
qf = df[mask].reset_index()

for idx, row in qf.iterrows():
    
    pts = props[idx].coords
    
    if fill_pass[idx]:
        plot_cell(ax, pts, lc='w', fc='grey')
    # else:
    #     plot_cell(ax, pts, color='red', lw=1)

plt.imshow(A, cmap='binary_r')    
```

```python

unsharp_radius = 75
unsharp_amount = 100
simple = False
min_peak_dist = 70
erode = 5
perc = 25 
blur = 3

dfList = []

for k, img in imgs.items():
    fid = "_".join(k.split("_")[:2])
    key = k.split("_")[-1]
    res, cells = segement_image(img, 
                                unshrp_rad=unsharp_radius,
                                unshrp_amnt=unsharp_amount,
                                erode=erode,
                                perc=perc,
                                blur=blur,
                                simple=simple)
    df, props = getFeatures(cells)
    
    df['key'] = key
    df['fid'] = fid
    dfList.append(df)
    
df = pd.concat(dfList, ignore_index=True)
print(df.shape)
print(df['key'].value_counts())
print(df.groupby('key')['area'].mean())

df.head()
```
