```python
import pandas as pd
import numpy as np
import os
import gget
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
import sys
import umap
import itertools
import scipy
import scanpy as sc

import utils as ut
```

```python
""" LOAD EMONT DATA """
fpath =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/WATAtlas/subtypes/preadipocyte_labels.parquet"
mdf = pd.read_parquet(fpath)
print(f"{mdf.shape=}")


fpath =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/WATAtlas/subtypes/preadipocyte_rna.parquet"
df = pd.read_parquet(fpath)
df = ut.normalize(df, 1e6)
print(f"{df.shape=}")
df.head()
```

```python
""" Load our data """

annPath =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/_global_clusters.csv"
ldf = pd.read_csv(annPath)

cellTypes = [
    'Stromal cells',
]

ldf = ldf[ldf['cellType'].isin(cellTypes)]
print(f"{ldf.shape=}")
print()
print(ldf['sampleInfo'].value_counts())

ldf.head()
 
```

```python
rnaPath =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/_global_rna.csv"
rdf = pd.read_csv(rnaPath)
rdf = rdf.rename(columns={'Unnamed: 0' : 'gene'})
rdf = rdf.set_index('gene')
rdf = rdf.T

# filter 
pdf = rdf[rdf.index.isin(ldf['cellId'])]
expGene = [x for x in pdf.columns if x in df.columns]
pdf = pdf[expGene]

pdf = ut.normalize(pdf, 1e6)

pdf.head()
```

```python
metric = 'correlation'

pdf = pdf[expGene]
df = df[expGene]

# normalize
pdf[expGene] = sklearn.preprocessing.minmax_scale( pdf[expGene], feature_range=(0, 1), axis=1)

df[expGene] = sklearn.preprocessing.minmax_scale(df[expGene], feature_range=(0, 1), axis=1)

dist = scipy.spatial.distance.cdist(pdf, df, metric)
print(f"{pdf.shape=}")
print(f"{df.shape=}")
print(f"{dist.shape=}")
```

```python
print(pdf.columns[0])
print(df.columns[0])
```

```python
pxf = pd.DataFrame(dist, index=pdf.index)

pxf = pd.merge(pxf, ldf[['cellId', 'sampleInfo']],
               how='left',
               left_index=True,
               right_on='cellId')

pxf = pd.melt(pxf, id_vars=['cellId', 'sampleInfo'])


pxf.head()


sns.boxplot(data=pxf, 
            x='sampleInfo',
            showfliers=False,
            y='value')
```
