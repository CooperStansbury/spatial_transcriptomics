```python
import os
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import scanpy as sc
from collections import Counter
import scipy
from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable
from sklearn import metrics
import gget
import tabulate
from sklearn.neighbors import BallTree
from scipy.stats import gmean


# locals
import utils as ut
```

```python
xyPath =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/combinedEmbedding.csv"

xy = pd.read_csv(xyPath)
xy.head()
```

```python
cardDir =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/"
keys = ['ND', 'HFD8', 'HFD14']

rna = {}
labels = {}

for key in keys:
    labelPath = f"{cardDir}{key}_macrophage_clusters.csv"
    lf = pd.read_csv(labelPath)
    
    lf = pd.merge(lf, xy[['x', 'y', 'cellId', 'colors']], 
                  how='left',
                  left_on='cellId',
                  right_on='cellId')
    
    labels[key] = lf
    
    
    rnaPath = f"{cardDir}{key}_macrophage_rna.csv"
    rf = pd.read_csv(rnaPath)
    rf = rf.T
    new_header = rf.iloc[0] 
    rf = rf[1:] 
    rf.columns = new_header 
    rf.index.names = ['cellId']
    
    rna[key] = rf

print('done')
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 14, 4
plt.rcParams['figure.facecolor'] = "w"


query = "Mki67"
cmap = 'copper'

for i, key in enumerate(keys):
    
    rf = rna[key]
    lf = labels[key]
    
    mask = xy['cellId'].isin(lf['cellId'])
    
    qmask = rf[query.upper()].astype(bool)
    
    print(f"{qmask.sum()} ({qmask.sum() / len(qmask):.4f}) {key} cells express {query} > 0")
    
    colors = rf[query.upper()][qmask]
    qmask = (xy['cellId'].isin(colors.index)) & (xy['key'] == key)
    
    fig, axs = plt.subplots(1, 3, gridspec_kw={'width_ratios': [3, 3, 1]})
    axs = axs.ravel()
    

    """ plot the cell types """
    p = axs[0].scatter(xy['x'][mask], 
                       xy['y'][mask], 
                       c=xy['colors'][mask], 
                       ec='k',
                       zorder=2,
                       lw=0.1,
                       marker=".",)
    
    p = axs[0].scatter(xy['x'][~mask], 
                       xy['y'][~mask], 
                       c='lightgrey', 
                       ec='None',
                       zorder=1,
                       lw=0.1,
                       marker=".",)
    

    """ plot the query on shared embedding """  
    u = axs[1].scatter(xy['x'][qmask], 
                       xy['y'][qmask], 
                       c=np.log(list(colors)), 
                       ec='k',
                       zorder=2,
                       cmap=cmap,
                       lw=0.1,
                       marker=".",)
    
    p = axs[1].scatter(xy['x'][~qmask], 
                       xy['y'][~qmask], 
                       c='lightgrey', 
                       ec='None',
                       zorder=1,
                       lw=0.1,
                       marker=".",)
    
    """Plot the proportion of cells positive for the query """
    gf = pd.DataFrame(xy[xy['key'] == key]['cellTypes'].value_counts()).reset_index()
    gf.columns = ['cellType', 'GlobalCount']
    
    bf = pd.DataFrame(xy[qmask]['cellTypes'].value_counts()).reset_index()
    bf.columns = ['cellType', 'QueryCount']
    
    bf = pd.merge(bf, gf, 
                  how='left', 
                  left_on='cellType', 
                  right_on='cellType')
    
    bf['QueryProp'] = bf['QueryCount'] / bf['QueryCount'].sum()
    bf['GlobalProp'] = bf['QueryCount'] / bf['GlobalCount']
    
    toMerge = xy[['cellTypes', 'colors']].drop_duplicates()
    
    bf = pd.merge(bf, toMerge, 
                  how='left', 
                  left_on='cellType', 
                  right_on='cellTypes')
    
    bf = bf.sort_values(by='cellType')
    bf = bf.set_index('cellType')
    colors = list(bf['colors'].unique())
    
    
    print(bf[['QueryCount', 
              'GlobalCount', 
              'QueryProp',
              'GlobalProp']])
    
    bf[['QueryProp']].T.plot(kind='bar', 
                             stacked=True, 
                             edgecolor='k', 
                             ax=axs[2],
                             color=colors)
    
    axs[2].set_xticks([], [])
    axs[2].yaxis.tick_right()
    axs[2].yaxis.set_label_position("right")
    axs[2].set_ylabel(f"Proportion")
    axs[2].set_title(f"{query.lower().capitalize()}+ Cells")
    axs[2].legend(bbox_to_anchor=(1.4, 1.015))

    plt.colorbar(u, ax=axs[1], label="Exp (log)")    
    axs[0].set_xlabel("")
    axs[0].set_ylabel("")
    axs[0].set_xticks([])
    axs[0].set_yticks([])
    axs[0].set_title(key)
    
    axs[1].set_xlabel("")
    axs[1].set_ylabel("")
    axs[1].set_xticks([])
    axs[1].set_yticks([])
    axs[1].set_title(query.lower().capitalize())

    
    plt.tight_layout() 
    plt.show()
    
```

```python
query = 'Mki67'
q = query.upper()

res = []

for i, key in enumerate(keys):
    
    rf = rna[key]
    lf = labels[key].drop_duplicates(subset='cellId')
    
    print(rf.shape, lf.shape)
    
    cTypes = list(sorted(lf['cellType'].unique()))
    
    for ctype in cTypes:
        cids = lf[lf['cellType'] == ctype]['cellId']
        mask = rf.index.isin(cids)
        
        inGroup = rf[mask]
        outGroup = rf[~mask]
        
        propExpress = inGroup[q].astype(bool).sum() / mask.sum()
        
        scores, pvals = stats.ttest_ind_from_stats(
            mean1=inGroup[q].mean(),
            std1=inGroup[q].std(),
            nobs1=inGroup.shape[0],
            mean2=outGroup[q].mean(),
            std2=outGroup[q].std(),
            nobs2=outGroup.shape[0],
            equal_var=False,  # Welch's
        )
        
        lfc = np.log(inGroup[q].mean()) - np.log(outGroup[q].mean())
        
        row = {
            'key' : key,
            'cellType' : ctype,
            'gene' : q,
            'pval' : pvals,
            'score' : scores,
            'LFC' : lfc,
            'groupMean' : inGroup[q].mean(),
            'otherMean' : outGroup[q].mean(),
            'propExpress' : propExpress,
        }
        
        res.append(row)
    
res = pd.DataFrame(res)
print(res.shape)
res.head()
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 12, 3
plt.rcParams['figure.facecolor'] = "w"

sns.barplot(data=res, 
            x='cellType', 
            y='propExpress', 
            hue='key', 
            edgecolor='k', 
            palette=['#e1eb28', '#368212', '#b36c1b'])

plt.legend(loc='upper right')
plt.xlabel("")
plt.ylabel(f'Proportion {q.lower().capitalize()}+')
```
