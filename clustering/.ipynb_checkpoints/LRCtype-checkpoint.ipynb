```python
cardDir =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/"
keys = ['ND', 'HFD8', 'HFD14']

rna = {}
labels = {}

for key in keys:
    labelPath = f"{cardDir}{key}_macrophage_clusters.csv"
    lf = pd.read_csv(labelPath)
    
    lf = pd.merge(lf, xy[['x', 'y', 'cellId', 'colors']], 
                  how='left',
                  left_on='cellId',
                  right_on='cellId')
    
    labels[key] = lf
    
    
    rnaPath = f"{cardDir}{key}_macrophage_rna.csv"
    rf = pd.read_csv(rnaPath)
    rf = rf.T
    new_header = rf.iloc[0] 
    rf = rf[1:] 
    rf.columns = new_header 
    rf.index.names = ['cellId']
    
    rna[key] = rf

print('done')
```

```python
n = 25
cmap = 'viridis'
criteria = 'precision'

cTypes = list(sorted(lig['cellType'].unique()))


for ctype in cTypes:
    
    plt.rcParams['figure.dpi'] = 300
    plt.rcParams['figure.figsize'] = 12, 5
    plt.rcParams['figure.facecolor'] = "w"
        
    fig, axs = plt.subplots(1, 3)
    axs = axs.ravel()
    
    kdf = lig[lig['cellType'] == ctype]

    for i, key in enumerate(keys):
        cf = kdf[kdf['key'] == key]
        cf = cf[~cf.isin([np.nan, np.inf, -np.inf]).any(1)]
        cf = cf.sort_values(by=criteria, ascending=False)
        
        sns.scatterplot(data=cf.head(n), 
                        x=criteria, 
                        y='gene', 
                        hue='LFC',
                        ec='k',
                        s=70,
                        palette=cmap,
                        zorder=2,
                        ax=axs[i])
        
        norm = plt.Normalize(cf[criteria].head(n).min(), 
                             cf[criteria].head(n).max())
        
        sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
        sm.set_array([])
        
        plt.colorbar(sm, ax=axs[i], label='Log Fold Change')
        
        axs[i].get_legend().remove()
        axs[i].set_ylabel("")
        axs[i].set_xlabel(f"{criteria.capitalize()}")
        axs[i].set_title(key)
        axs[i].grid(True, zorder=1)
        
    plt.suptitle(f"Ligands for {ctype}")
    plt.tight_layout()
    plt.show()

    break

        
```

```python
F1t = 0.6
alpha = 0.05
ctype = "Mac5"
time = "HFD14"
lrType = 'ligand'

# db = 'WikiPathways_2019_Mouse'
# db = 'KEGG_2019_Mouse'
db = 'ontology'

mask = (df['cellType'] == ctype) & (df['key'] == time) & (df['F1'] > F1t) & (df['type'] == lrType)

tdf = df[mask]
print(tdf.shape)
print(tdf['type'].value_counts())
print()

genes = list(tdf['gene'].unique())
ef = gget.enrichr(genes, database=db)
ef = ef[ef['p_val'] <= alpha]
pd.set_option('display.max_colwidth', 200)
ef[['path_name', 'overlapping_genes']].head(15)
```

```python
pathId = 0

record = ef.iloc[pathId]

print(f"==== {record['path_name']} ====")


res = []
for l in record['overlapping_genes']:
    
    lrpairs = lr[lr['ligand'] == l]
    
    for r in lrpairs['receptor']:
        mask = (df['key'] == 'HFD14') & (df['type'] != lrType) & (df['gene'] == r)
        kdf = df[mask]
        kdf = kdf.sort_values(by='F1', ascending=False)
        kdf['ligand'] = l
        kdf['receptor'] = r
        
        res.append(kdf[['cellType', 'ligand', 'receptor', 'F1', 'LFC']])

res = pd.concat(res, ignore_index=True)
print(res[['ligand', 'receptor']].value_counts().reset_index())

plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = 12, 3
plt.rcParams['figure.facecolor'] = "w"

sns.barplot(data=res, 
             x='cellType', 
             y='LFC',
             zorder=1,
             edgecolor='k',
             hue='ligand', 
             palette='Set3')

plt.grid(True, zorder=0)
plt.xlabel("")
plt.title(record['path_name'])

plt.legend(bbox_to_anchor=(1.01, 1.025), title="Receptors for")
_ = plt.xticks(rotation=45)
```

```python
Lctype = "Mac4"
Rctype = "Mac5"
lims = 5
metric = 'F1'
cols = [
    'cellType',
    'gene', 
    'sensitivity', 
    'specificity', 
    'precision', 
    'true_positive',
    'meanGroup',
    'totalCells',
    'F1',
    'LFC'
]

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 15, 5
plt.rcParams['figure.facecolor'] = "w"

fig, axs = plt.subplots(1, 3)
axs = axs.ravel()

df['totalCells'] = df[['true_positive', 'false_negative']].sum(axis=1)


lmask = (df['cellType'] == Lctype) & (df['type'] == 'ligand')
rmask = (df['cellType'] == Rctype) & (df['type'] == 'receptor')

ldf = df[lmask].reset_index(drop=True)
rdf = df[rmask].reset_index(drop=True)

for i, key in enumerate(keys):
    
    lk = ldf[ldf['key'] == key][cols]
    rk = rdf[rdf['key'] == key][cols]
    
    lk.columns = [f"l_{x}" for x in cols]
    rk.columns = [f"r_{x}" for x in cols]
        
    # print(lk.head())
    
    # filter the LR table
    lrt = lr[lr['ligand'].isin(lk['l_gene'])]
    lrt = lrt[lrt['receptor'].isin(rk['r_gene'])]
    
    lrt = lrt[['LR', 'ligand', 'receptor', 'Ligand.CC']].reset_index(drop=True)
    
    lrt = pd.merge(lrt, lk, 
                   left_on='ligand', 
                   right_on='l_gene')
    
    lrt = pd.merge(lrt, rk, 
                   left_on='receptor', 
                   right_on='r_gene')
    
    
    # lrt['color'] = (lrt['l_true_positive'] + lrt['r_true_positive']) / (lrt['l_totalCells'] + lrt['r_totalCells'])

    lrt['color'] = lrt[['l_F1', 'r_F1']].apply(lambda x: scipy.stats.hmean(x), axis=1)

    sns.scatterplot(data=lrt, 
                    x='l_LFC', 
                    y='r_LFC', 
                    ec='k',
                    zorder=2,
                    hue='color',
                    legend=False,
                    s=20,
                    palette='RdYlGn',
                    ax=axs[i])
    
    axs[i].grid(True, zorder=1)
    axs[i].set_ylabel(f"Ligand LFC")
    axs[i].set_xlabel(f"Receptor LFC")
    axs[i].set_title(f"{key}")
    axs[i].set_ylim([-lims, lims])
    axs[i].set_xlim([-lims, lims])
    
    n = len(lrt)
    nmask = (lrt['l_LFC'] > 0) & (lrt['r_LFC'] > 0)
    print(f"{key} {nmask.sum()} of {n} in blue ({nmask.sum() / n:.4f})")
    
    rect = patches.Rectangle((0, 0), 
                             lims, lims, 
                             linewidth=1, 
                             zorder=0,
                             facecolor='lightblue', 
                             alpha=0.5)

    # Add the patch to the Axes
    axs[i].add_patch(rect)
    norm = plt.Normalize(0, 1)
    sm = plt.cm.ScalarMappable(cmap='RdYlGn', norm=norm)
    sm.set_array([])
    plt.colorbar(sm, ax=axs[i], label='hmean F1')
    
plt.suptitle(f"{Lctype} Ligands vs. {Rctype} Receptors")
plt.tight_layout()
```
