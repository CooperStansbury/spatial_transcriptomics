```python
"""load scRNA"""
dirPath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/scanpy/"

rna = {}

for f in os.listdir(dirPath):
    
    if f.endswith('.h5ad'):
        key = f.split(".")[0]
        fullPath = f"{dirPath}{f}"
        data = sc.read(fullPath)
        rna[key] = data
    
print("done")
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 12, 4
plt.rcParams['figure.facecolor'] = "w"

k = 7

keys = ['ND', 'HFD8', 'HFD14']

res = []

for key in keys:
    data = rna[key]
    
    # get run key
    allKeys = list(data.obs.keys() )
    allKeys.remove('n_genes')
    runKey = allKeys[0]
    
    # 2D for plotting
    embedding = data.obsm['X_umap']

    # get original cluster Labels
    clusterIds = data.obs[runKey].astype(int) + 1 # note that these are zero-indexed!
    k = clusterIds.nunique()
    clusterIds = pd.DataFrame(clusterIds)
    clusterIds = clusterIds.reset_index(drop=False)
    clusterIds.columns = ['cellId', 'Leiden']
    
    # get distance matrix    
    A = data.obsp['distances'].todense()
    
    # print(f"{A.min()=} {A.max()=}")
    
    """Figure params"""
    alpha = 0.5
    s = 25
    edc = 'grey'
    cmap = 'tab20'
    fig, axs = plt.subplots(1, 3, sharey=True)
    axs = axs.ravel()
    
    
    """ Leiden algorithm """
    axs[0].scatter(embedding[:, 0], 
                   embedding[:, 1], 
                   c=clusterIds['Leiden'],
                   alpha=alpha,
                   s=s,
                   cmap=cmap,
                   edgecolor=edc)
    
    axs[0].set_aspect('equal')
    axs[0].set_xticks([])
    axs[0].set_yticks([])
    
    axs[0].set_ylabel('UMAP 2')
    axs[0].set_xlabel('UMAP 1')
    axs[0].set_title(f'{key} Leiden Clustering')
    
    """ Spectral clustering """
    # recluster using other methods
    spect = SpectralClustering(n_clusters=k,
                               affinity='precomputed',
                               assign_labels='discretize',
                               random_state=0).fit(A)
    
    clusterIds['Spectral'] = spect.labels_
    
    axs[1].scatter(embedding[:, 0], 
                   embedding[:, 1], 
                   c=clusterIds['Spectral'],
                   alpha=alpha,
                   s=s,
                   cmap=cmap,
                   edgecolor=edc)
    
    axs[1].set_aspect('equal')
    axs[1].set_xticks([])
    axs[1].set_yticks([])
    
    axs[1].set_xlabel('UMAP 1')
    axs[1].set_title(f'{key} Spectral Clustering')
    
    rand = metrics.rand_score(clusterIds['Leiden'], clusterIds['Spectral'])
    mi = metrics.adjusted_mutual_info_score(clusterIds['Leiden'], clusterIds['Spectral'])
    print(f'Leiden vs. Spectral Rand: {rand:.4f}, MI: {mi:.4f}')
    
    
    """kmeans"""
    
    kmeans = KMeans(n_clusters=k).fit(A)
    
    clusterIds['KMeans'] = kmeans.labels_
    
    axs[2].scatter(embedding[:, 0], 
                   embedding[:, 1], 
                   c=clusterIds['KMeans'],
                   alpha=alpha,
                   s=s,
                   cmap=cmap,
                   edgecolor=edc)
    
    axs[2].set_aspect('equal')
    axs[2].set_xticks([])
    axs[2].set_yticks([])
    
    axs[2].set_xlabel('UMAP 1')
    axs[2].set_title(f'{key} KMeans Clustering')
    
    rand = metrics.rand_score(clusterIds['Leiden'], clusterIds['KMeans'])
    mi = metrics.adjusted_mutual_info_score(clusterIds['Leiden'], clusterIds['KMeans'])
    print(f'Leiden vs. KMeans Rand: {rand:.4f}, MI: {mi:.4f}')
    
    plt.tight_layout()
    plt.show()
    
    clusterIds['key'] = key
    res.append(clusterIds)
    break
    
    
res = pd.concat(res, ignore_index=True)
res.head()
```
