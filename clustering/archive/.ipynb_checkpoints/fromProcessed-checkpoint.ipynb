```python

plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 6, 4

# cluster scoring 
krange = np.linspace(2, 17, 16).astype(int)
# nrange = np.linspace(2, n_components-1, 10).astype(int)
n = 10
newRows = []


for d, data in adata.items():
    print(f"working {d}")
    for k in krange:
        data.simpleClustering(k, n, use_umap=True)
        res = data.scoreClusters()
        res['k'] = k
        res['data'] = d
        newRows.append(res)
    
res = pd.DataFrame(newRows)

sns.lineplot(data=res,
             x='k',
             y='silhouette_score',
             hue='data',
             marker="o")

ax = plt.gca()
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[0:], labels=labels[0:])

_ = plt.xticks(krange, krange)
plt.ylabel("Silhouette Coefficient")
plt.xlabel("k")
```

```python

plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 6, 4

sns.lineplot(data=res,
             x='k',
             y='davies_bouldin_score',
             hue='data',
             marker="o")

ax = plt.gca()
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[0:], labels=labels[0:])

_ = plt.xticks(krange, krange)
plt.ylabel("Davies Bouldin Index")
plt.xlabel("k")
```

```python
k = 6
n = 120
c = ut.ncolor(k, cmap='viridis')

plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 6, 4

for d, data in adata.items():
    data.simpleClustering(k, n=n, use_umap=False)

    for i in range(k):
        # annotate all cells
        plt.scatter(data.umap_embedding[data.labels == i , 0] , 
                    data.umap_embedding[data.labels == i , 1] , 
                    marker=".",
                    label=i+1,
                    alpha=0.4,
                    c=c[i])
        
    plt.legend(bbox_to_anchor=(1.2, 1.025))
    plt.xlabel("UMAP 1")
    plt.ylabel("UMAP 2")
    plt.title(d)
    plt.show()
    
```
