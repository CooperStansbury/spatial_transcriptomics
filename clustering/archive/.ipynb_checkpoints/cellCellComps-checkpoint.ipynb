```python
clusterPath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/scanpy/clusters.csv"
cf = pd.read_parquet(clusterPath)
print(cf.shape)
print(cf['cellType'].unique())


"""subset cell type of interest """
alpha = 0.05
lft = 1
cellType = 'Macrophages'

keys = ['ND', 'HFD8', 'HFD14']

res = {}
allGenes = []
lfcGenes = []

for key in keys:

    t = cf[cf['cellType'] == cellType].reset_index(drop=True)
    t = t[t['key'] == key].reset_index(drop=True)
    nClusters = t['clusterId'].nunique()
    t['geneCount'] = t.groupby('gene')['clusterId'].transform('count')

    t = t[t['pvals'] < alpha] # only singificant results

    uni = t[t['geneCount'] < 2] # single cluster
    shared = t[t['geneCount'] > (nClusters-1)] # all clusters
    allGenes += shared['gene'].to_list()
    
    lfc = shared[shared['logfoldchanges'] > lft]
    lfcGenes += lfc['gene'].to_list()
    
    print(key)
    print(f"{uni.shape} {uni['gene'].nunique()}")
    print(f"{shared.shape} {shared['gene'].nunique()}")
    
    res[key] = [uni, shared]
    
    print()
    
    
# get all genes observed in all three time points
lfcGenes = [item for item, count in collections.Counter(lfcGenes).items() if count > 2]
allGenes = [item for item, count in collections.Counter(allGenes).items() if count > 2]

print(f"{len(allGenes)=}")
print(f"{len(lfcGenes)=}")
```

```python
dirPath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/scanpy/"

adata = {}

for f in os.listdir(dirPath):
    
    if f.endswith('.h5ad'):
        key = f.split(".")[0]
        fullPath = f"{dirPath}{f}"
        data = sc.read(fullPath)
        adata[key] = data
    
print("done")
```

```python
threshold = 0.5


plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 6, 6
plt.rcParams['figure.facecolor'] = "w"

metric = "correlation"

for key in keys:
    data = adata[key]
    
    df = data.to_df()
    # get the run key
    allKeys = list(data.obs.keys() )
    allKeys.remove('n_genes')
    runKey = allKeys[0]
    print(key, runKey)
    
    clusterIds = data.obs[runKey].astype(int) + 1  # note that these are zero-indexed!
    
    uni, shared = res[key]
    
    cids = shared['clusterId'].unique() 
    cells = clusterIds.isin(cids)
    cells = df[cells]
    cells = cells[lfcGenes]
    indexGenes = np.array(lfcGenes)
    
    D = squareform(pdist(cells.T, metric)) # symetric KL divergence between within cluster cells
    D *= 1.0/D.max()
    
    print(f"{D.mean()=}")
    
    meanDist = np.mean(D, axis=1)
    
    mask = (meanDist < threshold)
    differentGenes = indexGenes[mask]
    print(", ".join([f"'{x}'" for x in differentGenes]))
    print()
    print("\n".join(differentGenes))
    print()
    
     # sort by most similar
    ind = np.argsort(D.sum(axis=0))[::-1]
    Dhat = D[ind, :][:, ind]
    
    plt.matshow(Dhat)
    
    plt.colorbar(fraction=0.046, 
                 pad=0.04, 
                 label="1 - Correlation")
    
    _ = plt.xticks([])
    _ = plt.yticks([])
    
    plt.xlabel(f"{cellType} DEG")
    plt.ylabel(f"{cellType} DEG")
    plt.title(key)
    
    plt.show()
    


    # print(shared.head())
    
    # break
```

```python
clusterPath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/scanpy/clusters.csv"
cf = pd.read_parquet(clusterPath)
print(cf.shape)
print(cf['cellType'].unique())


"""subset cell type of interest """
alpha = 0.05
lft = 2
cellTypes = ['Macrophages', 'B cells']

keys = ['ND', 'HFD8', 'HFD14']

res = {}

for key in keys:
    res[key] = {}
    for ctype in cellTypes:
        t = cf[cf['cellType'] == ctype].reset_index(drop=True)
        t = t[t['key'] == key].reset_index(drop=True)
        nClusters = t['clusterId'].nunique()
        t['geneCount'] = t.groupby('gene')['clusterId'].transform('count')

        t = t[t['pvals'] < alpha] # only singificant results
        shared = t[t['geneCount'] > (nClusters-1)] # all clusters
        lfc = shared[shared['logfoldchanges'] > lft]
        res[key][ctype] = lfc['gene'].to_list()
    

    
sharedG = {}    
for time, cells in res.items():
    for cellType, genes in cells.items():
        if not cellType in sharedG:
            sharedG[cellType] = genes
        else:
            sharedG[cellType] += genes
        

for ct, genes in sharedG.items():
    lfcGenes = [item for item, count in collections.Counter(genes).items() if count > 2]
    sharedG[ct] = lfcGenes
    print(ct, len(lfcGenes))
```

```python
threshold = 0.2

metric = 'correlation'

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 6, 6
plt.rcParams['figure.facecolor'] = "w"

for key in keys:
    data = adata[key]
    
    df = data.to_df()
    # get the run key
    allKeys = list(data.obs.keys() )
    allKeys.remove('n_genes')
    runKey = allKeys[0]
    print(key, runKey)
    
    clusterIds = data.obs[runKey].astype(int) + 1  # note that these are zero-indexed!
    
    # get all genes
    allGenes = [x for v in sharedG.values() for x in v]

    dfList = []
    
    for ct in cellTypes:
        cids = cf[(cf['key'] == key) & (cf['cellType'] == ct)]['clusterId'].unique()
        cells = clusterIds.isin(cids)
        cells = df[cells].reset_index(drop=True)
        cells = cells[allGenes]
        dfList.append(cells)
        
    df = pd.concat(dfList, ignore_index=True)
    
    # D = squareform(pdist(df.T, jsd)) # symetric KL divergence between within cluster cells
    D = squareform(pdist(df.T, metric)) # symetric KL divergence between within cluster cells
    # D *= 1.0/D.max()
    print(D.shape)
    plt.matshow(D, vmin=0, vmax=1)
    
    plt.axvline(x=len(sharedG[cellTypes[0]])-0.5, c='k')
    plt.axhline(y=len(sharedG[cellTypes[0]])-0.5, c='k')
    plt.colorbar(fraction=0.046, 
                 pad=0.04,
                 label='1-Correlation')
             # label=str(metric).capitalize())
    
    l1 = (len(sharedG[cellTypes[0]])-0.5) / 2
    l2 = ((len(sharedG[cellTypes[1]])-0.5) / 2 ) + len(sharedG[cellTypes[0]], )
    
    _ = plt.xticks([l1, l2], [cellTypes[0], cellTypes[1]])
    _ = plt.yticks([])
    
    plt.title(key)
        
    plt.show()
    

 
```
