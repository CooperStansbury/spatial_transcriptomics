```python
# load cluster assignments
cdf = pd.read_csv("clusterAssignments.csv")
cid = {}

for key in cdf['key'].unique():
    t = cdf[cdf['key'] == key]
    cid[key] = t
    
print('done')
cdf['cellType'].unique()
```

```python
dirPath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/scanpy/"

adata = {}

for f in os.listdir(dirPath):
    
    if f.endswith('.h5ad'):
        key = f.split(".")[0]
        fullPath = f"{dirPath}{f}"
        data = sc.read(fullPath)
        adata[key] = data
    
print("done")
```

# global common genes

```python
# first, get genes present in all three times points

genesByKey = []
keys = ['ND', 'HFD8', 'HFD14']
for i, key in enumerate(keys):
    data = adata[key]    
    df = data.to_df()
    
#     # get the n genes with highest average expression
#     exp = df.std(axis=0)
#     ind = np.argsort(exp)
#     mask = (ind < nCells) # highest is first in argsort result
#     cf = df.T[mask].T
        
    genes = set(df.columns)
    print(key, len(genes))
    genesByKey.append(genes)
    

commmonGenes = set.intersection(*genesByKey)
print(f"{len(commmonGenes)} genes at all three times")
```

```python
topN = 20

ctypes = ['Monocytes', 'Macrophages']
keys = ['ND', 'HFD8', 'HFD14']

genesByType = {x: [] for x in ctypes}
genesByType


for key in keys:
    data = adata[key]
    result = data.uns['rank_genes_groups']
    keys = data.uns['rank_genes_groups'].keys()
    
    cellTypes = cid[key]
    
    for ctype in ctypes:
        cellTypeIds = cellTypes[cellTypes['cellType'] == ctype]['clusterId'].to_numpy()

        for ctid in cellTypeIds:
            genes = result['names'][str(ctid-1)][:topN] # remember that cluster IDs are zero-based here!
            [genesByType[ctype].append(x) for x in genes]
            

# make sure that each gene is expressed at each time point
for k, v in genesByType.items():
    vset = set(v)
    aset = vset.intersection(commmonGenes)
    genesByType[k] = aset
    print(k, len(aset))
```

```python
runKey = "nn15_pca30_r0.75"

keys = ['ND', 'HFD8', 'HFD14']
ctypes = ['Monocytes', 'Macrophages']
pltSize = 4
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.facecolor'] = 'w'

res = {}

for key in keys:
    data = adata[key]
    clusterIds = data.obs[runKey].astype(int) + 1  # note that these are zero-indexed!
    
    df = data.to_df()
    
    # get all clusters for the cell type
    cellTypes = cid[key]
    
    # set up the plot
    nPlots = len(ctypes)
    plt.rcParams['figure.figsize'] = nPlots*pltSize, pltSize
    fig, axs = plt.subplots(1, 
                            nPlots, 
                            subplot_kw=dict(box_aspect=1))
    axs = axs.ravel()
    
    res[key] = {}
    
    for i, ctype in enumerate(ctypes):     
        # get the cell-type specific genes
        cGenes = list(genesByType[ctype])
        cf = df[cGenes]
        
        cellTypeIds = cellTypes[cellTypes['cellType'] == ctype]['clusterId'].to_numpy()
        mask = clusterIds.isin(cellTypeIds)   
        
        # get cells with the desired label
        cf = cf[mask]
        genes = cf.columns
        
        # build gene-gene correlation using these cells
        X = np.corrcoef(cf.T)
        X = np.where(np.isnan(X), 0, X)
        np.fill_diagonal(X, 0)
        
        res[key][ctype] = X
        
        axs[i].matshow(X)
        axs[i].set_xticks([])
        axs[i].set_yticks([])
        axs[i].set_xlabel('Genes')
        
        if i == 0:
            axs[i].set_ylabel('Genes')
        axs[i].set_title(f"{key} {ctype}")
        
        
    plt.subplots_adjust(wspace=0.05)
    plt.show()
    
print('done')
```
