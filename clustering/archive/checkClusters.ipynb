```python
import os
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
import scanpy as sc
from sklearn import metrics
import tabulate
import itertools
from sklearn.decomposition import PCA
import matplotlib.patches as mpatches
import umap
from importlib import reload
from scipy import stats
import matplotlib.image as mpimg
import gget

# locals
import utils as ut
reload(ut)
```

```python
cleanDir = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/cleanData/"

labels = {}
rna = {}


for f in os.listdir(cleanDir):
    if "clusters" in f:
        key = f.split("_")[0]
        fpath = f"{cleanDir}{f}"
        lf = pd.read_parquet(fpath)
        lf['key'] = key
        lf['cellId'] = lf['cellId'].str.replace("-", ".")
        print(f"{key=}, {lf.shape=}")
        labels[key] = lf
        
    if "rna" in f:
        key = f.split("_")[0]
        fpath = f"{cleanDir}{f}"
        rf = pd.read_parquet(fpath)
        rf.index = rf.index.str.replace("-", ".")
        rf = ut.normalize(rf) # normalize the rna data
        print(f"{key=}, {rf.shape=}")
        rna[key] = rf
        
        
print("done")
```

```python
keys = ['ND', 'HFD8', 'HFD14']

PCAr = 2
UMAPr = 2

data = {}


for key in keys:
    lf = labels[key]
    rf = rna[key]
    
    # need to filter out the spots not considered (low-depth)
    rf = rf[rf.index.isin(lf['cellId'])]
        
    pca = PCA(n_components=PCAr)
    pcaModel = pca.fit(rf.T)
    Xpca = pd.DataFrame(pcaModel.components_.T, 
                        index=rf.index)
    
    pcaCols = [f"pc{x+1}" for x in range(PCAr)]
    Xpca.columns = pcaCols
    
    lf = pd.merge(lf, Xpca, 
                  how='left',
                  left_on='cellId', 
                  right_index=True)
    
    # plotting params
    n_neighbors = 50
    min_dist = 0.25
    metric= "euclidean"

    umap_reducer = umap.UMAP(n_components=UMAPr,
                             n_neighbors=n_neighbors,
                             min_dist=min_dist,
                             metric=metric)

    ef = umap_reducer.fit_transform(Xpca) # we need embedding by cell
    Xumap = pd.DataFrame(ef, index=rf.index)
    
    umapCols = [f"umap{x+1}" for x in range(UMAPr)]
    Xumap.columns = umapCols
    
    lf = pd.merge(lf, Xumap, 
                  how='left',
                  left_on='cellId', 
                  right_index=True)
    
    # get the cluster centroids in UMAP space: NOTE centroids are cluster specific
    centers = []
    for col in umapCols:
        center = f"center{col.replace('umap', '')}"
        centers.append(center)
        # lf[center] = lf.groupby(['cellType', 'clusterId'])[col].transform('mean')
        lf[center] = lf.groupby(['cellType'])[col].transform('mean')
    
    # get distances from center
    lf['coords'] = lf[umapCols].apply(lambda x: np.array(x), axis=1)
    lf['centers'] = lf[centers].apply(lambda x: np.array(x), axis=1)
    lf['dist'] = lf[['coords', 'centers']].apply(lambda x: np.linalg.norm(x[0] - x[1]), axis=1)
    
    data[key] = lf

print('done')
```

```python
alpha = 0.05
alpha_hat = 1 - (1 - alpha)**(1/(7*3)) # bonferronis for 3 time points, 7 cell types

res = []
misses = {}

for i, key in enumerate(keys):
    
    lf = data[key]
    
    colors = plt.cm.get_cmap('Spectral', lut=7)
    bins = np.linspace(0, 20, 40)
    
    mf = []
    
    for j, ctype in enumerate(sorted(lf['cellType'].unique())):
        
        cf = lf[lf['cellType'] == ctype].reset_index()
        zscores = (cf['dist'] - cf['dist'].mean()) / cf['dist'].std()
        px = np.asarray([st.norm.cdf(x) for x in zscores])
        mask = np.where(px > (1 - alpha_hat))
        
        
        possibleMisses = cf[cf.index.isin(mask[0])]        
        
        mf.append(possibleMisses)
        
        row = {
            'key' : key, 
            'cellType' : ctype,
            'alpha' : alpha,
            'alpha_hat' : alpha_hat,
            'n' : len(mask[0]),
            'prop' : len(mask[0]) / len(cf),
        }
        res.append(row)
        
    misses[key] = pd.concat(mf)
    
res = pd.DataFrame(res)
res.head()
```
