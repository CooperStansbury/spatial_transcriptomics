```python
cardDir =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/"
xyPath =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/combinedEmbedding.csv"
xy = pd.read_csv(xyPath)

print()

keys = ['ND', 'HFD8', 'HFD14']

rna = {}
labels = {}

for key in keys:
    print(f"{key}...")
    labelPath = f"{cardDir}{key}_macrophage_clusters.csv"
    lf = pd.read_csv(labelPath)
    
    lf = lf.drop_duplicates(subset='cellId')
    
    lf = pd.merge(lf, xy[['x', 'y', 'cellId', 'colors']], 
                  how='left',
                  left_on='cellId',
                  right_on='cellId')
    

    cellIds = lf['cellId'].to_list()
    labels[key] = lf
    
    rnaPath = f"{cardDir}{key}_macrophage_rna.csv"
    rf = pd.read_csv(rnaPath)
    rf = rf.T
    new_header = rf.iloc[0] 
    rf = rf[1:] 
    rf.columns = new_header 
    rf.index.names = ['cellId']

    rf = rf[rf.index.isin(cellIds)]
    print(f"{key=} {lf.shape=} {rf.shape}")
    
    rf = ut.normalize(rf, 1e6)
    rna[key] = rf
    

print('done')
```

```python
fPath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/LR/LewisLabUSCS/Mouse/Mouse-2020-Baccin-LR-pairs.xlsx"

lr = pd.read_excel(fPath, engine="openpyxl")
print(lr.shape)

lr['ligand'] = lr['Ligand.Mouse'].str.upper()
lr['receptor'] = lr['Receptor.Mouse'].str.upper()
lr.head()
```

```python
sptDir =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/"

spt = {}

for f in os.listdir(sptDir):
    if "spt" in f and "global" in f:
        print(f)
        fullPath = f"{sptDir}{f}"
        key = f.split("_")[0]
        sdf = pd.read_csv(fullPath)
        sdf = sdf.rename(columns={'Unnamed: 0' : 'gene'})
        sdf = sdf.set_index('gene')
        sdf = sdf.T
        sdf.index = sdf.index.str.replace("-", ".")
        sdf = ut.normalize(sdf, 1e6)
        print(f"{key=} {sdf.shape=}")
        spt[key] = sdf
    
print('done')
```

```python
# get all ligands and receptors

allLR = lr['ligand'].to_list() + lr['receptor'].to_list()
print(f"{len(allLR)=}")
allLR = list(set(allLR)) # filter out duplicates
allLR = [x for x in allLR if x in rna['ND'].columns] # filter out genes not expressed
print(f"{len(allLR)=}")
```

# get DEG for all cell types

```python
deg = []

macTypes = [
    'Mac1',
    'Mac2',
    'Mac3',
    'Mac4',
    'Mac5',
]


for key in keys:
    rf = rna[key][allLR]
    lf = labels[key]
    lf['metaType'] = np.where(lf['cellType'].isin(macTypes), 'Macrophages', lf['cellType'])

    for ct in lf['metaType'].unique():
        if ct == 'Stromal cells':
            continue
            
        print(f"{key} {ct}...")
        cdf = lf[lf['metaType'] == ct]
        mask = rf.index.isin(cdf['cellId'].to_list())

        inGroup = rf[mask]
        outGroup = rf[~mask]

        for g in allLR:
            score, pval = scipy.stats.ranksums(inGroup[g], outGroup[g],
                                           alternative='two-sided')

            lfc = np.log2(inGroup[g].mean()+1) - np.log2(outGroup[g].mean()+1)
            row = {
                'gene' : g, 
                'cellType' : ct,
                'key' : key,
                'pval' : pval, 
                'log2foldchange' : lfc
            }
            deg.append(row)


deg = pd.DataFrame(deg)
print(f"{deg.shape}")

deg.head()
```

```python
# filters
lfc = 1
alpha = 0.05
ntests = len(deg) / 7 # cell types
alphaHat = 1 - ((1-alpha) ** (1/ntests))
print(f"{alpha=} {ntests=} {alphaHat}")

sig = deg[deg['log2foldchange'] > lfc]
# sig = sig[sig['pval'] <= alphaHat]
print(f"{sig.shape}")
print()
sig['cellType'].value_counts()
```

# Check for colocalization

```python
# get LR pairs with sig genes
sigGenes = sig['gene'].unique()

lPair = lr[lr['ligand'].isin(sigGenes)]
lPair = lPair[~lPair['receptor'].str.contains("&")] # drop improoerly formatted pairs
rPair = lr[lr['receptor'].isin(sigGenes)]
rPair = rPair[~rPair['ligand'].str.contains("&")]

lrPair = pd.concat([lPair, rPair])
print(f'{lrPair.shape=}')
lrPair.head()
```

```python
coloc = []

for key in keys:
    sdf = spt[key]
    print(key, sdf.shape)

    for _, row in lrPair.iterrows():
        ligand = row['ligand']
        receptor = row['receptor']

        if not ligand in sdf.columns:
            continue

        if not receptor in sdf.columns:
            continue

        l = sdf[ligand]
        r = sdf[receptor]

        cor, pval = scipy.stats.pearsonr(l, r)
        co = np.sum([l.astype(bool), r.astype(bool)], axis=0)
        co = len(co[co > 1])

        row = {
            'ligand' : ligand,
            'receptor' : receptor,
            'key' : key,
            'correlation' : cor,
            'pval' : pval,
            'colocalization' : co,
            'coloc_percent' : co/sdf.shape[0]
        }

        coloc.append(row)

coloc = pd.DataFrame(coloc)
print(f"{coloc.shape}")

coloc.head()
```

```python
cdf  = coloc[coloc['colocalization'] > 0]
cdf['key'].value_counts()
```

```python
res = []

for idx, row in cdf.iterrows():

    key = row['key']
    l = row['ligand']
    r = row['receptor']

    # get ligand hits
    lsig = sig[(sig['gene'] == l) & (sig['key'] == key)]
    lsig = lsig.sort_values(by='log2foldchange', ascending=False)
    
    rsig = sig[(sig['gene'] == r) & (sig['key'] == key)]
    rsig = rsig.sort_values(by='log2foldchange', ascending=False)

    if len(lsig) < 1:
        continue

    if len(rsig) < 1:
        continue

    newRow = {
        'key' : key,
        'ligand' : l,
        'receptor' : r,
        'ligandCtype' : lsig['cellType'].values[0],
        'receptorCtype' : rsig['cellType'].values[0],
        'ligandLFC' : lsig['log2foldchange'].values[0],
        'receptorLFC' : rsig['log2foldchange'].values[0],
        'coloc_percent' : row['coloc_percent'],
        'nLig' : len(lsig),
        'nRec' : len(rsig),
    }

    res.append(newRow)

res = pd.DataFrame(res)
res.head()
```
