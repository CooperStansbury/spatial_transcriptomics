```python
def getKnn(kdf, k=6, metric='minkowski'):
    """A function to generate a knn graph """
    
    nodes = kdf[['x', 'y']]
      
    # get the n points around that node
    nbrs = NearestNeighbors(n_neighbors=k+1,  
                            metric=metric, 
                            algorithm='ball_tree').fit(nodes)
    distances, indices = nbrs.kneighbors(nodes)
    return indices


def knnInd2Id(kdf, knnInd):
    """Translate neighbors to spotIds """
    
    transId = {}
    
    for i, row in kdf.iterrows():
        spotId = row['spotId']
        nbrhs = kdf.iloc[knnInd[i]]['spotId'].values
        transId[spotId] = nbrhs
    return transId
    
    

def blur(sdf, transId):
    """A function to blur all columns using their neighbors.
    Expects spotIds as index """
    
    results  = []
    
    for spotId, nbhs in transId.items():
        
        t = sdf.loc[nbhs].mean().to_dict()
        t['spotId'] = spotId
        results.append(t)
        
    results = pd.DataFrame(results)
    results = results.set_index('spotId')
    return results
```

```python
pid = "mmu05010"
genes = ut.parseKEGG(pid)
genes = [x.upper() for x in genes if x.upper() in spt['ND'].columns]

pname = ut.getPathname(pid)
print(f"{pname=} {len(genes)=}")
```

```python
## build some tensors

rank = 2
knn = 6
sigma = 3

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 6, 26

fig, axs = plt.subplots(1, 3, sharey=True)
axs = axs.ravel()


for i, key in enumerate(['ND', 'HFD8', 'HFD14']):
    kdf = df[df['key'] == key].reset_index(drop=True)
    
    # make coords simpler, introduces some error
    kdf['x_int'] = kdf['x'].astype(int)
    kdf['y_int'] = kdf['y'].astype(int)
    
    coords = kdf[['spotId', 'x_int', 'y_int']]
    
    # get spatial data
    sdf = spt[key][set(genes)]
    print(f"{key} {sdf.shape=}")
    
    # blur the spatial data
    knnInd = getKnn(kdf, k=knn)
    k2Ind = knnInd2Id(kdf, knnInd)
    sdf = blur(sdf, k2Ind)
    
    
    # merge positions
    sdf = pd.merge(sdf, coords, 
                   how='left',
                   left_index=True,
                   right_on='spotId')
    
    
    H = []
    for g in set(genes):
        ldf = pd.pivot_table(sdf, 
                             index=['y_int'], 
                             columns=['x_int'],
                             # fill_value=0,
                             values=g)

        # binarize
        # ldf = np.where(ldf > t, 1, 0)

        yIndices = ldf.index.to_list()
        xIndices = ldf.columns.to_list()
        ldf = ldf.to_numpy()
        H.append(ldf)
    

    # construct the tensor
    H = np.asarray(H)
    tensor_mask = np.isfinite(H)
    tensor = tl.tensor(H)
    
    tensor = np.nan_to_num(tensor)
    
    # blur
    if not sigma is None:
        for j in range(len(tensor)):
            tensor[j, :, :] = gaussian_filter(tensor[j, :, :], sigma)
    

    # decompose the tensor
    factors = tl.decomposition.non_negative_parafac(tensor,
                                                    normalize_factors=True,
                                                    mask=tensor_mask,
                                                    rank=rank)
    weights, factors = factors
    
    # gene plots
    cdf = pd.DataFrame(factors[0], index=set(genes))
    cdf.columns = [f"CC {i+1}" for i in range(rank)]
    
    sns.heatmap(cdf,
                linewidths=0.2,
                # annot=True,
                # fmt=".1f",
                cbar=False,
                cmap='viridis',
                ax=axs[i])
    
    axs[i].set_title(key)
    
    # break
    
plt.tight_layout()
```
