```python
import sys
import numpy as np

sys.path.append("/home/cstansbu/.local/lib/python3.7/site-packages/pycircos/")
import pycircos
import collections
import matplotlib.pyplot as plt
Garc    = pycircos.Garc
Gcircle = pycircos.Gcircle
```

```python
#Set chromosomes
circle = Gcircle() 


for i in range(10):
    name   = f"LR{i}"
    length = 3
    arc    = Garc(arc_id=name, 
                  size=length,
                  interspace=3, 
                  raxis_range=(950,1000),
                  labelposition=60, 
                  label_visible=True)
    circle.add_garc(arc) 
    
circle.set_garcs() 
```
