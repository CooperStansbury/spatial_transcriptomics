```python
def gaus2d(x=0, y=0, mx=0, my=0, sx=1, sy=1):
    return 1. / (2. * np.pi * sx * sy) * np.exp(-((x - mx)**2. / (2. * sx**2.) + (y - my)**2. / (2. * sy**2.)))

def makeData(n, m, x, y, sx, sy):
    X = np.linspace(0, 1, m)
    Y = np.linspace(0, 1, n)
    X, Y = np.meshgrid(X, Y)
    
    z = gaus2d(X, Y, x, y, sx, sy)
    return z

    
img1 = makeData(10, 10, 0.1, 0.3, 0.1, 0.2)  
img2 = makeData(10, 10, 0.1, 0.3, 0.5, 0.4)  
img3 = makeData(10, 10, 0.8, 0.8, 0.1, 0.1)  

# get correlations
df = pd.DataFrame({
    'Gene 1' : img1.ravel(),
    'Gene 2' : img2.ravel(),
    'Gene 3' : img3.ravel(),
})

cor = df.corr()
evals, evec = np.linalg.eig(cor)
print(f"{evals=}")
m1 = (1/np.sqrt(3)) * np.std(evals)
print(f"{m1=:.4f}")


plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.facecolor'] = "w"
plt.rcParams['figure.figsize'] = 9, 6

fig, axs = plt.subplots(1, 4)
axs = axs.ravel()

sns.heatmap(img1, 
            ax=axs[0], 
            linewidths=1,
            cbar=False,
            square=True,
            cmap='viridis')

axs[0].set_yticks([])
axs[0].set_xticks([])
axs[0].set_ylabel("Spatial Y")
axs[0].set_xlabel("Spatial X")
axs[0].set_title("Gene 1")

sns.heatmap(img2, 
            ax=axs[1], 
            linewidths=1,
            cbar=False,
            square=True,
            cmap='viridis')

axs[1].set_yticks([])
axs[1].set_xticks([])
axs[1].set_ylabel("")
axs[1].set_xlabel("Spatial X")
axs[1].set_title("Gene 2")


sns.heatmap(img3, 
            ax=axs[2], 
            linewidths=1,
            cbar=False,
            square=True,
            cmap='viridis')

axs[2].set_yticks([])
axs[2].set_xticks([])
axs[2].set_ylabel("")
axs[2].set_xlabel("Spatial X")
axs[2].set_title("Gene 3")


sns.heatmap(cor, 
            ax=axs[3], 
            linewidths=1,
            vmin=-1,
            vmax=1,
            annot=True,
            cbar=False,
            square=True,
            cmap='RdYlGn')
axs[3].set_title("Spatial Correlation")

# plt.suptitle(f"Multi-correlation {m1:.3f}")

plt.tight_layout()
```

```python
sptDir =  "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/data/CARDInputs/"

spt = {}

for f in os.listdir(sptDir):
    if "spt" in f and "macrophage" in f:
        print(f)
        fullPath = f"{sptDir}{f}"
        key = f.split("_")[0]
        sdf = pd.read_csv(fullPath)
        sdf = sdf.rename(columns={'Unnamed: 0' : 'gene'})
        sdf = sdf.set_index('gene')
        sdf = sdf.T
        sdf.index = sdf.index.str.replace("-", ".")
        spt[key] = sdf
    
print('done')
```

```python
def parseKEGG(pathId):
    genes = []
    results = REST.kegg_get(pathId).read()
    current_section = None
    for line in results.rstrip().split("\n"):
        section = line[:12].strip()  # section names are within 12 columns
        if not section == "":
            current_section = section

        if current_section == "GENE":
            linesplit = line[12:].split("; ")
            gene_identifiers = linesplit[0]
            gene_id, gene_symbol = gene_identifiers.split()
    
            if not gene_symbol in genes:
                genes.append(gene_symbol)
    return genes

def getPathname(pathId):
    """A function to return the legg pathname"""
    result = REST.kegg_list(pathId).read()
    return result.split("\t")[1].split("-")[0].strip()


pathId = "04110"
pathId = f"mmu{pathId}"
pathname = getPathname(pathId)
genes = parseKEGG(pathId)
genes = [x.upper() for x in genes]
print(f"{pathname} - {len(genes)} Genes")
genes[:10]
```

```python
keys = ['ND', 'HFD8', 'HFD14']

sampleSize = 10000
order = 3

res = []

for key in keys:

    sdf = spt[key]
    expGenes = [x for x in genes if x in sdf.columns]
    print(f"{key} {len(expGenes)}")
    
    sdf = sdf[expGenes]
    
    X = sdf.corr()
    X = X.fillna(0)
    print(f"{X.shape=}")

    comb = list(itertools.combinations(X.columns.to_list(), order))
    print(f"{len(comb)} combination of {len(expGenes)} with order={order}")
    print(f"Sampling {sampleSize}...")
    
    
    geneCombs = np.random.choice(list(range(len(comb))), 
                                      sampleSize, replace=False)
    
    for ind in geneCombs:
        
        mutliway = comb[ind]
        
        Y = X[list(mutliway)].loc[list(mutliway)].to_numpy()
        np.fill_diagonal(Y, 1)
        
        # method 2
        r1 = np.sqrt(1 - np.linalg.det(Y))
        
        # method 3
        l, _ = np.linalg.eigh(Y)
        r2 = (1/np.sqrt(order)) * np.std(l)
        
        row = {
            'key' : key,
            'genes' : mutliway,
            'r1' : r1,
            'r2' : r2,
        }
        res.append(row)
    
res = pd.DataFrame(res)
res.head()
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.facecolor'] = "w"
plt.rcParams['figure.figsize'] = 8, 3

sns.kdeplot(data=res, 
             x='r1', 
             bw_adjust=5,
             common_norm=True, 
             alpha=.5,
             fill=True,
             hue='key')


plt.xlabel(f'{order}-way multi-correlation')
plt.title(f'Distribution of spatial multi-correlations')
```
